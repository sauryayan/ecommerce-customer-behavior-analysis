```python
# Handling Missing review ratings
# We will populate median review ratings for each category
df['Category'].value_counts()
```

```python
# Creating a dictionary of median ratings for each category
dict_review = {}
for cat in set(df['Category']): # Created a set to get the unique categories
    rating = df.loc[df['Category'] == cat]['Review Rating'].median()
    # print(f"Review rating for {cat} - {rating}")
    dict_review[cat] = rating

dict_review
```

```python
null_review_index = df[df['Review Rating'].isnull()]['Review Rating'].index
```

```python
# We can also do that by using transform on grouped dataframe
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# Checking whether we have any null value
df.isnull().sum()
```

```python
# We have column names with upper and lowercase along with special characters
# For the ease of use and workability we can do snake casing
# Means lowercase letters with _ in between 
```

```python
# Renamed Columns
for column in df.columns:
    df.rename(columns={column: column.lower().replace(' ', '_')}, inplace=True)

df
```

```python
# Renaming purchase_amount_(usd) to purchase_amount
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.head()
```

```python
# Doing some feature engineering and create a new column age_group
labels = ['Young Adults', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
df[['age', 'age_group']].head(10)
```

```python
# Creating column purchase_frequency_days
# Since frequency_of_purchases is in text format, so we need to map them based on numbers
# So that we can analyze the numeric data

frequency_mapping = {
    'Annually': 365,
    'Bi-Weekly': 14,
    'Every 3 Months': 90,
    'Fortnightly': 14,
    'Monthly': 30,
    'Quarterly': 90,
    'Weekly': 7
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
# Checking discount_applied and promo_code_used columes are same and are redundant
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
# Another way to check that
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo_code_used
df = df.drop(columns='promo_code_used')
```

```python
from sqlalchemy import create_engine
import os

# Step 1: Connect to PostgreSQL
db_user = os.environ.get('MY_DB_USER')
db_password = os.environ.get('MY_DB_PASSWORD')
db_host = os.environ.get('HOST')
db_port = os.environ.get('PORT')

username = db_user
password = db_password
host = db_host
port = db_port
database = 'customer_behavior'

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

# Step 2: Load Database into PostgreSQL
table_name = 'customer'
df.to_sql(table_name, engine, if_exists='replace', index=False)

print(f"Data Successfully loaded into table '{table_name}' in database '{database}'.")
```
